```python
# Write a Python Program to Check Prime Number?

num = int(input("Number:"))
if num > 1:
   for i in range(2, num):
     if (num % i)==0:
       print(num," is not a prime number")
       break
   else:
     print(num,"is a prime number")
```

```python
# Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

lower = 1
upper = 10000

print("prime numbers between",lower,"and",upper,"is:")

for num in range(lower, upper + 1):
    if num > 1:
       for i in range(2, num):
           if (num % i)==0:
               break
       else:
          print(num)
```

```python
# Write a Python Program to Check Armstrong Number?

num = int(input("Number:"))
sum = 0
temp = num

while temp > 0:
  digit = temp % 10
  sum   += digit**3
  temp //= 10

if num == sum:
  print(num,"is an armstrong number.")
else:
  print(num,"is not an armstrong number.")
```

```python
# Write a Python Program to Find Armstrong Number in an Interval?

lower = 10
upper = 10000

for num in range(lower, upper + 1):

   # order of number
   order = len(str(num))
    
   # initialize sum
   sum = 0

   temp = num
   while temp > 0:
       digit = temp % 10
       sum += digit ** order
       temp //= 10

   if num == sum:
       print(num)
```

```python
# Write a Python Program to Find LCM?

num1 = int(input("Num1: "))
num2 = int(input("num2: "))

max_num = max(num1, num2)

lcm = max_num

while True:
  if lcm % num1 ==0 and lcm % num2 ==0:
    break
  lcm += max_num

print("The LCM of",num1,"and",num2,"is",lcm)
```

```python
#  Write a Python Program to Find HCF?

num1 = int(input("Num1: "))
num2 = int(input("num2: "))

min_num = min(num1,num2)

hcf = 1

for i in range(1, min_num + 1):
  if num1 % i == 0 and num2 % i == 0:
     hcf = i

print("The HCF of",num1,"and",num2,"is",hcf)
```

```python
#  Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

# Prompt the user to enter a decimal number
decimal = int(input("Enter a decimal number: "))

# Convert decimal to binary
binary = bin(decimal)[2:]  # [2:] is used to remove the '0b' prefix

# Convert decimal to octal
octal = oct(decimal)[2:]  # [2:] is used to remove the '0o' prefix

# Convert decimal to hexadecimal
hexadecimal = hex(decimal)[2:]  # [2:] is used to remove the '0x' prefix

# Display the results
print("Binary:", binary)
print("Octal:", octal)
print("Hexadecimal:", hexadecimal)
```

```python
# Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

# Prompt the user to enter two numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Prompt the user to select an operation
print("Select an operation:")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")
operation = input("Enter operation number (1-4): ")

# Perform the selected operation
if operation == '1':
    result = num1 + num2
    operator = '+'
elif operation == '2':
    result = num1 - num2
    operator = '-'
elif operation == '3':
    result = num1 * num2
    operator = '*'
elif operation == '4':
    result = num1 / num2
    operator = '/'
else:
    print("Invalid operation!")
    exit()

# Display the result
print(f"{num1} {operator} {num2} = {result}")
```
